<H3><p style="font-family:Georgia;">Cartesian Function</H3>
<p style="font-family:Georgia;"> In a three-dimensional Cartesian coordinate system, a point is represented by an ordered triple of real numbers (x, y, z). The x-coordinate specifies the horizontal position of the point, the y-coordinate specifies the vertical position of the point, and the z-coordinate specifies the depth of the point. 
</p>
<p style="font-family:Georgia;">The Cartesian function formula for three sets of numbers is:</p>
<p style="font-family:Georgia;">p<sub>x,a</sub> = (R<sub>E</sub> + L<sub>a</sub>) cos(Î¸<sub>a</sub>) cos(Ï<sub>a</sub>)</p>
<p style="font-family:Georgia;">p<sub>y,a</sub> = (R<sub>E</sub> + L<sub>a</sub>) cos(Î¸<sub>a</sub>) sin(Ï<sub>a</sub>)</p>
<p style="font-family:Georgia;">p<sub>z,a</sub> = (R<sub>E</sub> + L<sub>a</sub>) sin(Î¸<sub>a</sub>) 

```python
#Creating Cartesian Function to convert the Latitude, Longitude and Altitude values to Cartesian form
def cartesian_conv(Lo, La, Al):
  R = 6378137
  X = (R + Al) * math.cos(La) * math.cos(Lo)
  Y =  (R + Al) * math.cos(La) * math.sin(Lo)
  Z =  (R + Al) * math.sin(La)  
  return (X,Y,Z)
```

```python
#Getting the cartesian distance in the XYZ Columns
for index, row in df.iterrows():
    X,Y,Z = cartesian_conv(row['Longitude'],row['Latitude'],row['Altitude'])

    df['X'].iloc[index]=X
    df['Y'].iloc[index]=Y
    df['Z'].iloc[index]=Z
```

<H5><p style="font-family:Georgia;">ED_GS1_to_A</H5>
<p style="font-family:Georgia;">Function called ED_GS1_to_A that calculates the Euclidean distance between a point represented by variables XGS1, YGS1, and ZGS1 and another point represented by variables X, Y, and Z. The function takes a single argument A, which is the index of the second point in a dataframe df.
The function begins by extracting the values of X, Y, and Z from the dataframe using the index A. It then calculates the Euclidean distance between the two points using the formula:
<p style="font-family:Georgia;">d = â((X<sub>GS1</sub> - X)<sup>2</sup> + (Y<sub>GS1</sub> - Y)<sup>2</sup> + (Z<sub>GS1</sub> - Z)<sup>2</sup>)

<p style="font-family:Georgia;">Finally, the function returns the calculated distance d as the output.</p>

<p style="font-family:Georgia;"><B>Function to calculate end to end transmission rate</B>
<p style="font-family:Georgia;">Code defines a function called tr_rate that takes a list of indices list_path as an argument and returns a rate value. The function calculates the rate by iterating through the indices in list_path, calculating the Euclidean distance between each pair of indices using the Euclidean_dis function, and determining the rate based on the distance calculated.
<p style="font-family:Georgia;">The function begins by initializing the variable best_rate to 0 and the variable GS to the last element in the list_path list. It then iterates through the indices in list_path, excluding the first and last indices, using a for loop. For each index, it calculates the distance between the current index and the next index using the Euclidean_dis function.
<p style="font-family:Georgia;">Based on the calculated distance, the function determines a rate value using a series of if-elif statements. If the distance is greater than 500,000, the rate is set to 31.895. If the distance is between 400,000 and 500,000, the rate is set to 43.505, and so on.
<p style="font-family:Georgia;">The function then compares the rate value to the current value of best_rate and updates best_rate if the current rate is lower. This process continues until all indices have been processed.
<p style="font-family:Georgia;">Finally, the function calculates the distance between the second-to-last index in list_path and one of the two fixed points (either "GS1" or "GS2") using the ED_GS1_to_A or ED_GS2_to_A function, depending on the value of GS. It then determines a rate value based on this distance using a series of if-elif statements, similar to the ones used earlier.

```python
#function to calculate end to end transmission rate
def tr_rate(list_path):
    best_rate=0
    GS = list_path[-1]
    for i in range(len(list_path)-2):
        d = Euclidean_dis(list_path[i],list_path[i+1])
        if d > 500000:
            r=31.895
        elif 500000 > d > 400000:
            r = 43.505
        elif 400000> d > 300000:
            r = 52.857
        elif 300000 > d > 190000:
            r = 63.970
        elif 190000 > d > 90000:
            r = 77.071
        elif 90000 > d > 35000:
            r = 93.854
        elif 35000 > d:
            r = 119.130
        if i == 0:
            best_rate = r
        elif r < best_rate : 
            best_rate = r

    if GS == "GS1":
        dis = ED_GS1_to_A(list_path[-2]) 
    elif GS == "GS2" : 
        dis = ED_GS2_to_A(list_path[-2]) 
    if dis > 500000:
        r = 31.895
    elif 500000 > dis > 400000:
        r = 43.505
    elif 400000 > dis > 300000:
        r = 52.857
    elif 300000 > dis > 190000:
        r = 63.970
    elif 190000 > dis > 90000:
        r = 77.071
    elif 90000 > dis > 35000:
        r = 93.854
    elif 35000 > dis :
        r = 119.130 
    if r < best_rate or best_rate == 0: 
            best_rate = r
    return(best_rate)
```

<p style="font-family:Georgia;"><B>Route a population to routes</B>
<p style="font-family:Georgia;">Function dividingroutes that takes a list of flight indices flightlist and a string GS as arguments and returns a list of routes.
<p style="font-family:Georgia;">The function iterates through the flights in flightlist and for each flight, it calculates the Euclidean distance between the last flight in the current route (represented by the list single_route) and the current flight, as well as the distance between the last flight in the current route and one of the two fixed points ("GS1" or "GS2"). It then compares the distances and decides whether to add the current flight to the current route or to start a new route. If the distance between the current flight and the last flight in the current route is less than the distance between the last flight in the current route and the fixed point, the current flight is added to the current route. If the distance between the current flight and the last flight in the current route is greater than or equal to the distance between the last flight in the current route and the fixed point, the current route is terminated and a new route is started.

```python
# Route a population to routes
def dividingroutes(flightlist, GS):
    list_of_routes = []
    single_route = []
    
    # looping through all flights to be added to list
    for flight in flightlist:
        if len(single_route) == 0:
            single_route.append(flight)
        else:
            last_flight = single_route[-1]
            
            # find distnace between flights
            
            d1 = Euclidean_dis(last_flight, flight)
            
            # find distnace between flight and GS
            if GS == 'GS1':   
                d2 = ED_GS1_to_A(last_flight)
            elif GS == 'GS2':
                d2 = ED_GS2_to_A(last_flight)
            
            # compare the distances
            #print(d1,d2), and (d2-d1) > 350000
            if d1 < d2 :
                single_route.append(flight)
            else:
                single_route.append(GS)
                list_of_routes.append(single_route)
                single_route = []
                
    return list_of_routes    
```

<p style="font-family:Georgia;">The function first initializes a variable d to be the Euclidean distance between the flight and the first flight in the flight_list. It also initializes a variable nextf to be the first flight in the flight_list.
<p style="font-family:Georgia;">Then, the function iterates through the rest of the flights in flight_list, starting from the second flight. For each of these flights, it calculates the Euclidean distance between the flight and the current flight and compares it to the current value of d. If the distance is smaller than d, then the function updates d to be the new distance and updates nextf to be the current flight.
<p style="font-family:Georgia;">Finally, the function returns nextf, which is the flight in flight_list that is closest to the flight according to the Euclidean distance.

```python
# Find next flight
def nextflight(flight, flight_list):
    d = Euclidean_dis(flight, flight_list[0])
    nextf = flight_list[0]
    for f in flight_list[1:]:
        d1 = Euclidean_dis(flight, flight_list[0])
        if d1 < d:
            d = d1
            nextf = f
    return nextf
```

<p style="font-family:Georgia;">The function begins by initializing two variables: 'GS and node'. If point1 is in the list of ground stations ("GS1" or "GS2"), then GS is set to point1 and node is set to point2. If point2 is in the list of ground stations, then GS is set to point2 and node is set to point1. If neither point1 nor point2 are in the list of ground stations, then GS is set to None.
<p style="font-family:Georgia;">If GS is not None, then the function calculates the distance between the ground station and node using one of two functions: ED_GS1_to_A or ED_GS2_to_A. If GS is None, then the function calculates the Euclidean distance between point1 and point2 using a function called Euclidean_dis.
<p style="font-family:Georgia;">After calculating the distance, the function uses a series of if statements to determine the data rate r based on the distance. If the distance is greater than 500000, then r is set to 31.895. If the distance is between 500000 and 400000, then r is set to 43.505, and so on. Finally, the function returns r.

```python
def datarate2points(point1, point2):
    GS_list = ["GS1", "GS2"]
    GS = None
    if point1 in GS_list:
        GS = point1
        node = point2
    elif point2 in GS_list:
        GS = point2
        node = point1
    
    if GS != None:
        if GS == "GS1":
            d = ED_GS1_to_A(node) 
        elif GS == "GS2" : 
            d = ED_GS2_to_A(node) 
    else:
        d = Euclidean_dis(point1, point2)
        
    if d > 500000:
        r=31.895
    elif 500000 > d > 400000:
        r = 43.505
    elif 400000> d > 300000:
        r = 52.857
    elif 300000 > d > 190000:
        r = 63.970
    elif 190000 > d > 90000:
        r = 77.071
    elif 90000 > d > 35000:
        r = 93.854
    elif 35000 > d :
        r = 119.130
    return r
```

<p style="font-family:Georgia;">The function begins by initializing an empty list called routes which will be used to store all routing paths. It also initializes a variable called currentfl to be equal to GS, a list called visited_list to store the flights that have already been visited, and a variable route which will be used to store the current routing path.

<p style="font-family:Georgia;">The function enters a loop that continues until currentfl becomes None. Inside the loop, the function checks whether currentfl is equal to GS. If it is, it selects the first flight in flight_list that has not been visited and sets currentfl to be this flight. It also initializes route to be a list containing only this flight and adds the flight to the visited_list. If all flights in flight_list have been visited, the function sets currentfl to be None to exit the loop.

<p style="font-family:Georgia;">If currentfl is not GS, then the function calculates the data transmission rate for the current routing path route plus the ground station GS. It then finds the next flight with the maximum data transmission rate from the current location using the nextflight function, which returns the next closest flight in flight_list that has not yet been visited. The function also calculates the data rate between the current flight and the next flight using the datarate2points function.

<p style="font-family:Georgia;">If the data rate between the current flight and the next flight is equal to the data transmission rate for the current routing path, then the function adds the next flight to the current routing path and marks it as visited. If the data rate is not equal to the data transmission rate, then the function adds the ground station to the current routing path, appends the routing path to the routes list, and sets the current flight back to the ground station.

<p style="font-family:Georgia;">Finally, the function returns the list of all routing paths, routes.

```python
def greedy1(flight_list, GS):
    # Create an empty list to store all routing paths
    routes = []
    currentfl = GS
    visited_list = []
    # While the current location is not a GS
    while currentfl != None:
        if currentfl == GS:
            currentfl = list(set(flight_list) - set(visited_list))[0]
            route = [currentfl]
            visited_list.append(currentfl)
            if len(flight_list) == len(visited_list):
                    currentfl = None
        else:
            # Find the next flight with the maximum data transmission rate from the current location
            routerate = tr_rate(route + [GS])
            nextfl = nextflight(currentfl, list(set(flight_list) - set(visited_list)))
            nextdrate = datarate2points(currentfl, nextfl)
            
            # only add the flight if it maintains the datarate
            if nextdrate == routerate:
                route.append(nextfl)
                visited_list.append(nextfl)
                currentfl = nextfl
                if len(flight_list) == len(visited_list):
                    currentfl = None

            else: 
                route.append(GS)
                routes.append(route)
                currentfl = GS
    return routes
```

<H5 style="color:teal;"><p style="font-family:Georgia;"><B>Fitness function</B></p></H5>
<p style="font-family:Georgia;">The fitness function evaluates an individualâs level of fitness (the ability of an individual to compete with other individuals). Each individual receives a fitness rating from the system. Based on its fitness score, an individual's likelihood of being chosen for reproduction is determined.
Our Fitness assessment function determines the data transmission rate and one path from point of initialization in order to solve the problem of maximising the end-to-end transmission rate, which is directly proportional to the minimum distance between two flights.<p style="font-family:Georgia;">

```python
# Fitness Function for Single Objective Optimization
def fit_fn_SO(link,GS):
    
    list_of_routes = dividingroutes(link,GS)
    list_of_rate = []
    best_rate=0
    for list_path in list_of_routes:

        for i in range(len(list_path)-2):
            d = Euclidean_dis(list_path[i],list_path[i+1])

            if d > 500000:
                r=31.895
            elif 500000 > d > 400000:
                r = 43.505
            elif 400000> d > 300000:
                r = 52.857
            elif 300000 > d > 190000:
                r = 63.970
            elif 190000 > d > 90000:
                r = 77.071
            elif 90000 > d > 35000:
                r = 93.854
            elif 35000 > d :
                r = 119.130
            if i == 0:
                best_rate = r
            elif r < best_rate : 
                best_rate = r
   
        if GS == "GS1":
            dis = ED_GS1_to_A(list_path[-2]) 
        elif GS == "GS2": 
            dis = ED_GS2_to_A(list_path[-2]) 
            
        if dis > 500000:
            r=31.895
        elif 500000 > dis > 400000:
            r = 43.505
        elif 400000> dis > 300000:
            r = 52.857
        elif 300000 > dis > 190000:
            r = 63.970
        elif 190000 > dis > 90000:
            r = 77.071
        elif 90000 > dis > 35000:
            r = 93.854
        elif 35000 > dis:
            r = 119.130 
        if r < best_rate or best_rate == 0 : 
                best_rate = r
        
        list_of_rate.append(best_rate)
        best_rate = 0
    
    average_rate = sum(list_of_rate)/len(list_of_rate)
   
    return(list_of_routes, average_rate)
```

<H5 style="color:teal;"><p style="font-family:Georgia;"><B>Weighted sum Method</B></p></H5>
<p style="font-family:Georgia;">The weighted sum method is a technique used to evaluate and compare multiple options or alternatives based on multiple criteria. It involves assigning weights to the criteria, calculating a score for each alternative based on its performance on the criteria, and summing the scores to obtain a overall score for each alternative.
<p style="font-family:Georgia;">The formula for the weighted sum method is:
<p style="font-family:Georgia;">Overall score = â(Weight<sub>i</sub> Ã Score<sub>i</sub>)
<p style="font-family:Georgia;">where:
<p style="font-family:Georgia;">Overall score is the overall score for the alternative being evaluated.
<p style="font-family:Georgia;">Weight<sub>i</sub> is the weight assigned to criterion i.
<p style="font-family:Georgia;">Score<sub>i</sub> is the score for the alternative on criterion i.
<p style="font-family:Georgia;">The overall score is calculated by summing the products of the weights and scores for each criterion. The weights reflect the relative importance of the criteria in the decision-making process, and the scores reflect the performance of the alternative on each criterion.


```python
# Fitness Function for Multiple Objective Optimization
def fit_fn_MO(link,GS):
    
    list_of_routes = dividingroutes(link,GS)
    list_of_rate = []
    list_of_latency = []
    best_rate=0
    for list_path in list_of_routes:

        for i in range(len(list_path)-2):
            d = Euclidean_dis(list_path[i],list_path[i+1])
            if d > 500000:
                r=31.895
            elif 500000 > d > 400000:
                r = 43.505
            elif 400000> d > 300000:
                r = 52.857
            elif 300000 > d > 190000:
                r = 63.970
            elif 190000 > d > 90000:
                r = 77.071
            elif 90000 > d > 35000:
                r = 93.854
            elif 35000 > d:
                r = 119.130
            if i == 0:
                best_rate = r
            elif r < best_rate : 
                best_rate = r
                
        if GS == "GS1":
            dis = ED_GS1_to_A(list_path[-2]) 
        elif GS == "GS2": 
            dis = ED_GS2_to_A(list_path[-2]) 
        if dis > 500000:
            r=31.895
        elif 500000 > dis > 400000:
            r = 43.505
        elif 400000> dis > 300000:
            r = 52.857
        elif 300000 > dis > 190000:
            r = 63.970
        elif 190000 > dis > 90000:
            r = 77.071
        elif 90000 > dis > 35000:
            r = 93.854
        elif 35000 > dis:
            r = 119.130 
        if r < best_rate or best_rate == 0: 
                best_rate = r
        
        list_of_rate.append(best_rate)
        best_rate = 0
        
        latency = 50 * (len(list_path)-1)
        list_of_latency.append(latency)
    
    average_rate = sum(list_of_rate)/len(list_of_rate)
    average_latency = sum(list_of_latency)/len(list_of_rate)
    
    # Weighted sum method 
    fitness = 0.5 * average_rate + 0.5 * average_latency
        
    return(list_of_routes, fitness)
```

<H5 style="color:teal;"><p style="font-family:Georgia;"><B>Tournament Selection</B></p></H5>
<p style="font-family:Georgia;">The purpose of the selection phase is to choose the most fit people and allow them to pass on their genes to the following generation. Based on their fitness ratings, two sets of people (parents) are chosen. High fitness people are more likely to be chosen for reproduction.
The tournament selection approach, where the best or fittest flights are chosen as parents and compete against one random chosen Ground Station.<p style="font-family:Georgia;">

```python
# Tournament Selection
def tour_sel(population,GS):
    parents = []
    num_of_parents = round(len(population)/2)
    if num_of_parents % 2 != 0:
        num_of_parents += 1
    copying_pop = copy.copy(population)
    for z in range(0,num_of_parents):
        p1,p2=random.sample(copying_pop,2)
        fitness_p1 =fit_fn_SO(p1,GS)
        fitness_p2 =fit_fn_SO(p2,GS)
        
        if fitness_p1[1]>fitness_p2[1]:
            parents.append(p1)
            copying_pop.remove(p1)
        else :
            parents.append(p2)
            copying_pop.remove(p2)
    return parents
```

<H5 style="color:teal;"><p style="font-family:Georgia;"><B>Two Point Crossover</B></p></H5>
<p style="font-family:Georgia;">The important phase of a genetic algorithm is crossover. A crossover point is picked at random from the genes for each set of parents to mate.
Two-point cross-over selection method with partial mapping is an efficient way of selecting genes for recombination in genetic algorithms. It is advantageous over other methods because it helps reduce the complexity of the selection process by using a simpler algorithm. The method also helps reduce the chances of introducing undesirable traits into the population and can help reduce the time taken to find a solution. Additionally, it also helps speed up the convergence of the population towards a more optimal solution.</p>

```python
# Two-point Cross-over Selection Method with partial mapping

def crossover(population):
    
    children = []
    for i in range(0,len(population),2):
        # get the parents 
        p1 = population[i]
        p2 = population[i+1]
        
        # chose two cross over points
        r1 = random.randint(2, int(0.75 * len(p1)))
        r2 = random.randint(r1, len(p1)-2)
        
        # give the corresponding points to child
        c1 = p1[:r1] + p2[r1:r2] + p1[r2:]
        c2 = p2[:r1] + p1[r1:r2] + p2[r2:]
        
        # partial mapping to avoid repitiotion
        for i in range(len(c1)):
            if  i < r1 or i >= r2:       
                f = c1[i]
                while f in c1[r1:r2]:
                    for g in c1[r1:r2]:
                        if g == f:    
                            ind = c1[r1:r2].index(g) + r1
                            break
                    f = c2[ind]
                c1[i] = f
                
        for i in range(len(c2)):
            if  i < r1 or i >= r2:       
                f = c2[i]
                while f in c2[r1:r2]:
                    for g in c2[r1:r2]:
                        if g == f:    
                            ind = c2[r1:r2].index(g) + r1
                            break
                    f = c1[ind]
                c2[i] = f
                
        children.append(c1)
        children.append(c2)
    return(children)
```

<p style="font-family:Georgia;">Scramble mutation is a type of genetic algorithm mutation which randomly reorders a section of the chromosome. It is a useful technique for searching a large number of solutions in a short amount of time. It can be used to solve complex problems, such as optimization and scheduling. Additionally, it is beneficial for use in multi-objective optimization, as it allows for a wider range of solutions to be explored.</p>

```python
# Scramble Mutation

def scramble_mut(population):
    
    scrambled =[]
    for f in population:
        r1 = random.randint(0, int(0.75 * len(f)))
        r2 = random.randint(r1, len(f))
        
        scramble = f[r1:r2]
        random.shuffle(scramble)
        s = f[:r1] + scramble + f[r2:]
        
        scrambled.append(s)
    return(scrambled)
```

<H5 style="color:brown;"><p style="font-family:Georgia;"> Implementation of Praticle Swarm Optimization 

<H5 style="color:black;"><p style="font-family:Georgia;">Single objective optimization</H5>
<H5 style="color:navy;"><p style="font-family:Georgia;">Intitalize Particles</H5>
<p style="font-family:Georgia;"> Defines a function called pso that takes a list of flight indices, an integer representing the size of the population, a string representing the fixed point, an integer representing the number of generations to run the particle swarm optimization algorithm, and two real numbers as arguments. The function does not return anything.

<p style="font-family:Georgia;"> The function initializes a list of particles and calculates the fitness of each particle using a helper function. It then compares the fitness of each particle with the current global best fitness and updates the global best if necessary. The global best is a list that stores the particle with the highest fitness and its fitness. The function also stores the global best fitness in a list and prints the initial population fitness.

<p style="font-family:Georgia;">A loop that iterates through the particles and generates a new velocity for each particle based on its previous velocity and the global best particle. The velocity is a list of swap pairs, where each swap pair represents a pair of indices that will be swapped in the particle.
<H5 style="color:Navy;"><p style="font-family:Georgia;">Find the personal & global fitness </H5>
<p style="font-family:Georgia;">The code iterates through the indices of the particle and checks if the flight at the current index is different from the flight at the same index in the personal best particle (the particle with the highest fitness among all the particles that the current particle has had in its history). If the flights are different, a swap pair is added to the velocity list.
<H5 style="color:Navy;"><p style="font-family:Georgia;">Velocity & position update for pBest</H5>
<p style="font-family:Georgia;">After the velocity list is created, the code iterates through the swap pairs in the velocity list and performs a random swap for each pair with a probability equal to 1 - beta. This means that the swap will be performed with a probability of beta and not performed with a probability of 1 - beta.
For each generation, the code iterates through the particles and generates a new velocity for each particle based on its previous velocity and the global best particle. The velocity is a list of swap pairs, where each swap pair represents a pair of indices that will be swapped in the particle.
<p style="font-family:Georgia;">The code then calculates the fitness of the particle using a helper function and updates the personal best and global best if necessary. The personal best is a list that stores the particle with the highest fitness among all the particles that the current particle has had in its history, and the global best is a list that stores the particle with the highest fitness among all the particles.
<p style="font-family:Georgia;">After all the particles have been updated, the code appends the global best fitness to a list and prints the fitness of the current generation.
<p style="font-family:Georgia;">Finally, the function returns the global best fitness, the routes of the global best particle, and the list of global best fitnesses for each generation.

```python
# particle swarm single objective optimization algorithm

def pso(flight_list, pop_size, GS, generation, alpha, beta):
    
    # Initialize particles
    particles = pop_int(flight_list,pop_size)
        
    # Find the personal & global fitness 
    pBest = []
    fitnessvsgen = []
    gBest = []
    for p in particles:
        routes, fitness = fit_fn_SO(p,GS)
        pBest.append([p, fitness])
        if len(gBest) == 0:
            gBest = [p, fitness]
        elif gBest[1] < fitness:
            gBest = [p,fitness,routes]

    fitnessvsgen.append(gBest[1])
    print(f'initial population fitness: {gBest[1]}')

    # create generations
    for i in range(generation):
        
        for j in range(len(particles)):
            particle = particles[j]
            pBest_particle = pBest[j][0]
            pBest_fitness = pBest[j][1]
            gBest_particle = gBest[0]
            gBest_fitness = gBest[1]

            # Velocity & position update for pBest

            velocity = []
            for k in range(len(particle)):
                if particle[k] != pBest_particle[k]: 
                    # create a swap pair
                    velocity.append((k, pBest_particle.index(particle[k])))
            
            for swap in velocity:
                ran = random.random()
                if ran > beta:
                    particle[swap[0]],particle[swap[1]] =particle[swap[1]],particle[swap[0]]
                    
            # Velocity & position update for gBest
            velocity = []
            for k in range(len(particle)):
                if particle[k] != gBest_particle[k]: 
                    # create a swap op
                    velocity.append((k, gBest_particle.index(particle[k])))
            
            for swap in velocity:
                ran = random.random()
                if ran > alpha:
                    particle[swap[0]],particle[swap[1]] =particle[swap[1]],particle[swap[0]]
                

            # Find the personal & global fitness  
            routes, fitness = fit_fn_SO(particle,GS)

            if fitness > pBest_fitness:
                pBest[j] = [particle, fitness]

            if fitness > gBest_fitness:
                gBest = [particle, fitness, routes]

        
        fitnessvsgen.append(gBest[1])
        print(f'{i + 1} generation fitness: {gBest[1]}')


    return(gBest[1], gBest[2], fitnessvsgen)
    
```

```python
import json
def JSONfile(Routes, filename):
    filename = filename + ".json"
    for route in Routes:
        routing_path = {"Source Node": None,
                            "Routing Path": [],
                            "End-to-end data rate": None}
        for i in range(len(route)):
            if i == 0:
                routing_path["Source Node"] = route[i]
                pre_node = route[i]
            else:
                rate = datarate2points(pre_node,route[i])
                routing_path["Routing Path"].append((route[i],rate))
                pre_node = route[i]
        routing_path["End-to-end data rate"] = tr_rate(route)
                
        # Serializing json
        json_object = json.dumps(routing_path, indent=2)
        # Writing to Routing_paths.json
        with open(filename, "a") as outfile:
            outfile.write(json_object)
            outfile.write('\n')
```
